```python
mine = pd.read_csv('mine.csv')['target']
mine_1 = pd.read_csv('mine_1.csv')['target']
add = pd.read_csv('add.csv')['target']
superb = pd.read_csv('super_blend.csv')['target']
clustering = pd.read_csv('clustering.csv')['target']
pca = pd.read_csv('pca_blender.csv')['target']
last =pd.read_csv('answer_14_my_scores_blended.csv')['target']
```

```python
solution_set = pd.DataFrame({'mine':mine,'mine_1':mine_1,'add':add,'superb':superb,'clustering':clustering,'pca':pca,'last':last})
```

```python
kendall = solution_set.corr(method='kendall')
```

```python
sns.heatmap(kendall,fmt='.3f',annot=True)
```

```python
submission = pd.read_csv('add.csv')
```
